```python
from sqlite3 import connect
```

```python
connection = connect("issue_tracker.db")
```

```python
einkaufswagen = connection.execute("""
create table if not exists issue (
                                   id integer primary key autoincrement,
                                   name text, 
                                   priority text,
                                   deadline datetime,
                                   is_done bool,
                                   description text
                                   );
""")
connection.commit()   
```

```python
def add_issue(name, description, deadline):
    connection = connect("issue_tracker.db")    
    sql = f"insert into issue (name, description, deadline, is_done) values ('{name}','{description}','{deadline}', 0);"
    einkaufswagen = connection.execute(sql)
    connection.commit()
    einkaufswagen.close()
    connection.close()

def get_issues(): 
    connection = connect("issue_tracker.db")    
    sql = "select * from issue where is_done = 0;"
    einkaufswagen = connection.execute(sql)
    data = einkaufswagen.fetchall()
    einkaufswagen.close()
    connection.close()
    return data 
    
def set_issue_to_done(issue_id):
    connection = connect("issue_tracker.db")    
    sql = f"update issue set is_done = 1 where id = {issue_id};"
    cursor = connection.execute(sql)
    connection.commit()
    cursor.close()
    connection.close()
```

```python
add_issue("Einkaufen gehen", "Kaufe Zahnpasta und Duschgel", "2026-05-01")
```
