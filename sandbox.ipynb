```python
from sqlite3 import connect
```

```python
connection = connect("issue_tracker.db")
```

```python
einkaufswagen = connection.execute("""
create table if not exists issue (
                                   id integer primary key autoincrement,
                                   name text, 
                                   priority text,
                                   deadline datetime,
                                   is_done bool,
                                   description text
                                   );
""")
connection.commit()   
```

```python
einkaufswagen = connection.execute("""
insert into issue (name, description, deadline) values 
                                   ('Einkaufen gehen', 'Kaufe Eier , Milch und Brot' ,'2026-04-22');
""")
connection.commit()   
```

```python
einkaufswagen= connection.execute(""" 
select * from issue;
""")
einkaufswagen.fetchall()
```

```python
def add_issue(name, description, deadline):
    connection = connect("issue_tracker.db")    
    sql = f"insert into issue (name, description, deadline) values ('{name}','{description}','{deadline}');"
    einkaufswagen = connection.execute(sql)
    connection.commit()
    einkaufswagen.close()
    connection.close()
```

```python
add_issue("Blumen kaufen", "Kaufe Blumen für die Mutter", "2026-05-01")
```
